```python
#check for outlires 
plt.figure(figsize=(18,15))
sns.boxplot(df[num_col])
plt.show()
```

<p> .We note that there is some columns have some outlires , let's check it out deeply</p>

```python
for col in num_col:
    Q1 = df[col].quantile(.25)
    Q3 = df[col].quantile(.75)
    IQR = Q3 - Q1
    Lower_Fence = Q1 - 1.5 * IQR
    Upper_Fence = Q3 + 1.5 * IQR
    Lower_Outliers = df[df[col] < Lower_Fence][col].values
    Upper_Outliers = df[df[col] > Upper_Fence][col].values
    df[col].replace(Lower_Outliers, Lower_Fence, inplace=True)
    df[col].replace(Upper_Outliers, Upper_Fence, inplace=True)
```

### Split Data

```python
from sklearn.model_selection import train_test_split
x= df.drop("deposit",axis=1)
y=df['deposit']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=0)
```

### Normlization Data

```python
from sklearn.preprocessing import MinMaxScaler
scalr = MinMaxScaler()
x_train[num_col]=scalr.fit_transform(x_train[num_col])
x_test[num_col]=scalr.transform(x_test[num_col])
```

### Encoding

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
ordinal = ['month','education']
label=['job','contact','loan','housing','marital']
ord=OrdinalEncoder()
lab=LabelEncoder()
x_train[ordinal]=ord.fit_transform(x_train[ordinal])
x_test[ordinal]=ord.transform(x_test[ordinal])
x_train[label]=ord.fit_transform(x_train[label])
x_test[label]=ord.transform(x_test[label])
x_train.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
yp=rfc.predict(x_test)
```

```python
accuracy_score(y_test,yp)*100
```
